```python
#don't run this! It takes a lot of time

# Obtain Info for Dataframe
bar = progressbar.ProgressBar()
gameMode = []
mapName = []
attacker_name = []
attacker_health = []
attacker_location_x = []
attacker_location_y = []
attacker_location_z = []
damage_reason = []
for m in bar(range(len(sample_matches))):
    curr_match = sample_matches[m]
    match = api.matches().get(curr_match)
    game_mode = match.game_mode
    map_name = match.map_name
    asset = match.assets[0]
    telemtry = api.telemetry(asset.url)
    armor_destroy_events = telemetry.events_from_type('LogArmorDestroy')
    for ade in armor_destroy_events:
        gameMode.append(game_mode)
        mapName.append(map_name)
        attacker = ade.attacker
        attacker_name.append(attacker.name)
        attacker_health.append(attacker.health)
        attacker_location_x.append(attacker.location.x)
        attacker_location_y.append(attacker.location.y)
        attacker_location_z.append(attacker.location.z)
        damage_reason.append(ade.damage_reason)
    
```

```python
#don't run this!
# Initialize Dataframe
data = {'Game_Mode':gameMode, 
        'Map_Name': mapName, 
        'Attacker_Name' : attacker_name, 
        'Attacker_Health': attacker_health, 
        'Attacker_location_x':attacker_location_x, 
        'Attacker_location_y':attacker_location_y, 
        'Attacker_location_z':attacker_location_z, 
        'Damage_Reason': damage_reason} 
```

```python
#don't run this!
df = pd.DataFrame(data)
len(df)
```
